```python
url="https://www.bikewale.com/royalenfield-bikes/"
```

```python
print(soup.title.text)
```

```python
print(soup.text)
```

```python
images=soup.find_all('img')
```

```python
print(images)
```

```python
for image in images:
    print(image['src'])
```
